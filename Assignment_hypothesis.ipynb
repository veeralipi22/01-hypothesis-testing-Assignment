```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import norm
```

```python
#Question 1
```

```python
#load the dataset
data = pd.read_csv('Cutlets.csv')
```

```python
data.head()
```

```python
UnitA=pd.Series(data.iloc[:,0])
UnitB=pd.Series(data.iloc[:,1])
```

```python
print(UnitA,UnitB)
```

```python
# 2-sample 2-tail ttest:   stats.ttest_ind(array1,array2)     # ind -> independent samples
p_value= stats.ttest_ind(UnitA,UnitB)
```

```python
p_value
```

```python
'''The p-value we got is 0.47223947245995,
which is greater than the significance level of 0.05.
When the p-value is greater than alpha, it suggests weak evidence against the null hypothesis.
In such cases, you would typically fail to reject the null hypothesis.'''
```
